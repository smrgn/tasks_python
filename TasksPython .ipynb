### Решение задачи №1

```python
def task1(z, y, x):
    a = ((x**2 + (z**3) / 75)**4 - y**3) / 15
    b = 8 * (1 - z**3)**4 + 8 * math.log(87 * x + 58 * y**2)
    c = 68 * (79 * x**2 + 1)**2 - (y - z**2)
    d = (x**3 - y**2)**7 + 77 * (93 * y**2 + 92 * z)
    return math.sqrt(a / b) - c / d

task1(0.86, -0.43, 0.2)
```

```python
def task2(z):
    if z < 155:
        return 60 * z**5 - 1 - z**2
    elif 155 <= z < 255:
        return 38 * z**6 + z**7
    elif 255 <= z < 323:
        a = (z**2 - 38 * z)**7
        b = 26 * (86 + 39 * z**2)**6
        c = (z**2 + 54 * z**3)**3
        return a + b + c
    elif 323 <= z < 388:
        return math.atan(z)**5 + (z - z**3 - z**2)**4 + 84 * math.log10(z)**6
    else:
        return 55 + (66 + 16 * z**2 + 56 * z)**6 / 65

task2(130)
```

```python
def task3(a, m, n, x):
    res_mult = 1
    res = 0
    for k in range(1, n + 1):
        for j in range(1, m + 1):
            for c in range(1, a + 1):
                res_mult *= (58 * (x**2 - 8 * c - j**3)**4 - math.sqrt(k)**6)
            res += res_mult
            res_mult = 1
    return res

task3(3, 5, 8, -0.74)
```

### Решение задачи №9

```python
def yes_or_no(string):
    if string == '0':
        return 'Нет'
    elif string == '1':
        return 'Да'
    return string


def transform_date(string):
    pattern = r'(\d+)\/(\d+)\/(\d+)'
    matche = re.findall(pattern, string)
    if len(matche) > 0:
        new_date = ''
        new_date += matche[0][-1][2:]
        new_date += '-'
        new_date += matche[0][1]
        new_date += '-'
        new_date += matche[0][0]
        return new_date
    return string


def transform_email(string):
    if '@' in string:
        string = string.replace('@', '[at]')
    return string


def transform_name(string):
    pattern = r'([А-Яа-яЁё]+),\s*([А-Я].)[А-Я].'
    matche = re.findall(pattern, string)
    if len(matche) > 0:
        new_name = ''
        new_name += matche[0][-1]
        new_name += ' '
        new_name += matche[0][0]
        return new_name
    return string


def F(string):
    string = transform_name(string)
    string = transform_email(string)
    string = yes_or_no(string)
    string = transform_date(string)
    return string


def task9(table):
    index = []
    for i in range(len(table)):
        temp = []
        for x in table[i]:
            if x is not None and x not in temp:
                temp.append(x)
        table[i] = temp
        if len(table[i]) != 0:
            for j in range(len(table[i])):
                table[i][j] = F(table[i][j])
        else:
            index.append(i)
    for j in index:
        table.pop(j)
    table_t = list(map(list, zip(*table)))
    return table_t
```

### Решение задачи №10

```python
class StateMachine:
    def __init__(self):
        self.state = "A"

    def etch(self):
        if self.state == "A":
            self.state = "B"
            return 0
        if self.state == "B":
            self.state = "C"
            return 2
        if self.state == "D":
            self.state = "E"
            return 4
        if self.state == "F":
            self.state = "B"
            return 8
        if self.state == "G":
            self.state = "B"
            return 9
        raise MealyError("etch")

    def pull(self):
        if self.state == "A":
            self.state = "E"
            return 1
        if self.state == "E":
            self.state = "F"
            return 5
        if self.state == "F":
            self.state = "F"
            return 7
        raise MealyError("pull")

    def spin(self):
        if self.state == "C":
            self.state = "D"
            return 3
        if self.state == "F":
            self.state = "G"
            return 6
        raise MealyError("spin")


class MealyError(Exception):
    pass


def main():
    return StateMachine()


def raises(method, error):
    output = None
    try:
        output = method()
    except Exception as e:
        assert type(e) == error
    assert output is None


def test():
    o = main()
    assert o.pull() == 1
    assert o.pull() == 5
    assert o.pull() == 7
    assert o.spin() == 6
    assert o.etch() == 9
    assert o.etch() == 2
    assert o.spin() == 3
    assert o.etch() == 4
    assert o.pull() == 5
    assert o.etch() == 8
    o = main()
    assert o.etch() == 0
    raises(lambda: o.pull(), MealyError)
    raises(lambda: o.spin(), MealyError)
    assert o.etch() == 2
    raises(lambda: o.etch(), MealyError)
```

### Решение задачи №11

```python
class Types:
    uint16 = 'H'
    int16 = 'h'
    int32 = 'i'
    uint32 = 'I'
    int8 = 'b'
    double = 'd'
    float_ = 'f'


class BinaryReader:
    def __init__(self, data, offset, order='<'):
        self.data = data
        self.offset = offset
        self.order = order

    def jump_to(self, offset):
        reader = BinaryReader(self.data, offset, self.order)
        return reader

    def read(self, frmt):
        data = unpack_from(self.order + frmt, self.data, self.offset)
        self.offset += calcsize(frmt)
        return data[0]


def read_d(reader):
    d1 = [reader.read(Types.int8) for _ in range(4)]
    d2_size = reader.read(Types.uint16)
    d2_offset = reader.read(Types.uint32)
    d2_reader = reader.jump_to(d2_offset)
    d2 = [d2_reader.read(Types.int8) for _ in range(d2_size)]
    d3 = reader.read(Types.float_)
    d4 = reader.read(Types.float_)
    return dict(D1=d1, D2=d2, D3=d3, D4=d4)


def read_c(reader):
    c1_size = reader.read(Types.uint32)
    c1_offset = reader.read(Types.uint32)
    c1_reader = reader.jump_to(c1_offset)
    c1 = [read_d(c1_reader) for _ in range(c1_size)]
    c2 = [reader.read(Types.uint16) for _ in range(6)]
    c3 = reader.read(Types.int16)
    return dict(C1=c1, C2=c2, C3=c3)


def read_b(reader):
    b1 = reader.read(Types.double)
    b2 = reader.read(Types.int8)
    return dict(B1=b1, B2=b2)


def read_a(reader):
    a1 = reader.read(Types.int16)
    a2 = reader.read(Types.int32)
    b_offset = reader.read(Types.uint32)
    b_reader = reader.jump_to(b_offset)
    a3 = read_b(b_reader)
    a4 = [reader.read(Types.int8) for _ in range(3)]
    a5 = reader.read(Types.int16)
    a6 = reader.read(Types.int8)
    a7 = read_c(reader)
    return dict(A1=a1, A2=a2, A3=a3, A4=a4, A5=a5, A6=a6, A7=a7)


def task11(data):
    return read_a(BinaryReader(data, 5))

task11(b'\x1cDZFE^{6\xde\xf4\xa1+\x00\x00\x00\x99\xbez\xb1\x81\x14\x02\x00\x00'
 b'\x00=\x00\x00\x00\xbd\x1b\xaa\x93F\x85.\x8aE\x8b\x8fV\xb4\t\xd8\xb9_\x1a\xa0'
 b'\x9f\xd0\xbf&\x89R\t%\x08[%.\xcdTZY\xbf\x06\x004\x00\x00\x001\xbbS?\xdd'
 b'\n\x8f>\xf8\xc5\xb0m\x03\x00:\x00\x00\x00\x0c\x02\x9a>\\`\x10\xbc')
```
