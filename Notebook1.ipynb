```python
import math as m
print("–ù–æ–ª—å –≤ –∫–∞—á–µ—Å—Ç–≤–µ –∑–Ω–∞–∫–∞ –æ–ø–µ—Ä–∞—Ü–∏–∏ –∑–∞–≤–µ—Ä—à–∏—Ç —Ä–∞–±–æ—Ç—É –ø—Ä–æ–≥—Ä–∞–º–º—ã")

while True:
    s = input("–í–≤–µ–¥–∏—Ç–µ –∑–Ω–∞–∫ (+, -, *, /, sin, cos, e^(x+y), x^y): ")
    if s == '0':
        break
    if s in ('+', '-', '*', '/', 'sin','cos', 'e^(x+y)', 'x^y'):
        x = float(input("x = "))
        y = float(input("y = "))
        if s == 'sin':
            print (m.sin(x+y))
        elif s == 'cos':
            print (m.cos(x+y))
        elif s == 'e^(x+y)':
            print (m.e**(x+y))
        elif s == 'x^y':
            print(x**y)
        elif s == '+':
            print(x + y)
        elif s == '-':
            print(x - y)
        elif s == '*':
            print(x * y)
        elif s == '/':
            if y != 0:
                print(x / y)
            else:
                print("–î–µ–ª–µ–Ω–∏–µ –Ω–∞ –Ω–æ–ª—å!")
    else:
        print("–ù–µ–≤–µ—Ä–Ω—ã–π –∑–Ω–∞–∫ –æ–ø–µ—Ä–∞—Ü–∏–∏. –ü–æ–≤—Ç–æ—Ä–∏—Ç–µ –≤–≤–æ–¥.")
```
